```python
# basic dependencies

import numpy as np
from numpy import loadtxt
from numpy import savetxt

import pandas as pd
import math
import time

np.set_printoptions(formatter={'float': lambda x: "{0:0.3f}".format(x)})

###########

# plotting dependencies
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline

# this is for the colorbar, you can change the cmap if you prefer other colour schemes
from matplotlib.cm import ScalarMappable
cm = plt.cm.get_cmap('viridis')

# function to return the std dev across runs
def ci(y, N_TRIALS):
    return 1.96 * y.std(axis=0) / np.sqrt(N_TRIALS)
```

```python
problem = "ZDT1"
max_hv = 0.8714093689206742

hvs_nsga3_all0 = loadtxt(f"{problem}_hvs_nsga3_dim8_200by2.csv", delimiter=',')
hvs_nsga3_all1 = loadtxt(f"{problem}_hvs_nsga3_dim8_100by4.csv", delimiter=',')
hvs_nsga3_all2 = loadtxt(f"{problem}_hvs_nsga3_dim8_50by8.csv", delimiter=',')
hvs_nsga3_all3 = loadtxt(f"{problem}_hvs_nsga3_dim8_25by16.csv", delimiter=',')

hv_nsga3_plot_ZDT1_0 = np.log10(max_hv - np.asarray(hvs_nsga3_all0)[...,:96]) 
hv_nsga3_plot_ZDT1_1 = np.log10(max_hv - np.asarray(hvs_nsga3_all1)[...,:48]) 
hv_nsga3_plot_ZDT1_2 = np.log10(max_hv - np.asarray(hvs_nsga3_all2)[...,:24]) 
hv_nsga3_plot_ZDT1_3 = np.log10(max_hv - np.asarray(hvs_nsga3_all3)[...,:12]) 

####

hvs_qnehvi_all0 = loadtxt(f"{problem}_hvs_qnehvi_dim8_100by2_1to10.csv", delimiter=',')

hvs_qnehvi_all1 = loadtxt(f"{problem}_hvs_qnehvi_dim8_50by4_1to10.csv", delimiter=',')

hvs_qnehvi_all2 = loadtxt(f"{problem}_hvs_qnehvi_dim8_25by8_1to10.csv", delimiter=',')

hv_qnehvi_plot_ZDT1_0 = np.log10(max_hv - np.asarray(hvs_qnehvi_all0)[...,:96]) 
hv_qnehvi_plot_ZDT1_1 = np.log10(max_hv - np.asarray(hvs_qnehvi_all1)[...,:48]) 
hv_qnehvi_plot_ZDT1_2 = np.log10(max_hv - np.asarray(hvs_qnehvi_all2)[...,:24]) 
```

```python
nsga3_dict = {}
nsga3_dict["ZDT1"] = [hv_nsga3_plot_ZDT1_0,hv_nsga3_plot_ZDT1_1,hv_nsga3_plot_ZDT1_2,hv_nsga3_plot_ZDT1_3]
nsga3_dict["C2-DTLZ2"] = [hv_nsga3_plot_LZ2_0,hv_nsga3_plot_LZ2_1,hv_nsga3_plot_LZ2_2,hv_nsga3_plot_LZ2_3]
nsga3_dict["MW7"] = [hv_nsga3_plot_MW7_0,hv_nsga3_plot_MW7_1,hv_nsga3_plot_MW7_2,hv_nsga3_plot_MW7_3]

qnehvi_dict = {}
qnehvi_dict["ZDT1"] = [hv_qnehvi_plot_ZDT1_0, hv_qnehvi_plot_ZDT1_1, hv_qnehvi_plot_ZDT1_2]
qnehvi_dict["C2-DTLZ2"] = [hv_qnehvi_plot_LZ2_0, hv_qnehvi_plot_LZ2_1, hv_qnehvi_plot_LZ2_2]
qnehvi_dict["MW7"] = [hv_qnehvi_plot_MW7_0, hv_qnehvi_plot_MW7_1, hv_qnehvi_plot_MW7_2]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(21, 5))

for i, problem in enumerate(["ZDT1", "C2-DTLZ2", "MW7"]):
    
    hvlist = nsga3_dict[problem]
    iterlist = [96, 48, 24, 12]
    batchlist = [2, 4, 8, 16]
    
    for hv, iteration, batch in zip(hvlist, iterlist, batchlist):    

        ax[i].plot(np.arange(iteration),
                hv.mean(axis=0),
                label=f'NSGA-III ({iteration} x {batch})', linewidth=1.5, alpha=0.7)

        ax[i].fill_between(np.arange(iteration),
                        hv.mean(axis=0)-ci(hv, 10), hv.mean(axis=0)+ci(hv, 10),
                        alpha=0.3)

    hvlist = qnehvi_dict[problem]
    iterlist = [96, 48, 24]
    batchlist = [2, 4, 8] 

    for hv, iteration, batch in zip(hvlist, iterlist, batchlist):    

        ax[i].plot(np.arange(iteration),
                hv.mean(axis=0),
                label=f'qNEHVI ({iteration} x {batch})', linewidth=1.5, alpha=0.7)

        ax[i].fill_between(np.arange(iteration),
                        hv.mean(axis=0)-ci(hv, 10), hv.mean(axis=0)+ci(hv, 10),
                        alpha=0.3)
        
    ax[i].set(xlabel='Iterations', ylabel='Log Hypervolume Difference')
    ax[i].grid(True)
    ax[i].set_title(f"{problem} (d=8), noise=0%, 10 runs")
    
ax[0].legend(loc='best')
```
